```python
import pandas as pd
```

```python
PollutionData = 'Resources/pollution_2000_2023.csv'
pollution_df = pd.read_csv(PollutionData, low_memory=False)
```

```python
pollution_df.head()
```

```python
#Column names & row count
print(pollution_df.columns)
print(len(pollution_df))
```

```python
#County names & county count
county_list = []
for row in pollution_df["County"]:
    if row not in county_list:
        county_list.append(row)
print(county_list)
print(len(county_list))
```
