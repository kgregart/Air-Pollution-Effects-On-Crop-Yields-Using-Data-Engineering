```python
import pandas as pd
```

```python
import os
path = 'Resources_TF'
extension = '.csv'
```

```python
files = [file for file in os.listdir(path) if file.endswith(extension)]
```

```python
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dfs.append(df)
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
clean_df = df.drop(columns=['1st Max Non Overlapping Value',
 '1st NO Max DateTime',
 '2nd Max Non Overlapping Value',
 '2nd NO Max DateTime','Date of Last Change'])
```

```python
from pydantic import BaseModel
from datetime import datetime

class Data(BaseModel):
    State_Code: int
    County_Code: int
    Site_Num: int
    Parameter_Code: int
    POC: int
    Latitude: float
    Longitude: float
    Datum: str
    Parameter_Name: str
    Sample_Duration: str
    Pollutant_Standard: str
    Metric_Used: str
    Method_Name: str
    Year: int
    Units_of_Measure: str
    Event_Type: str
    Observation_Count: int
    Observation_Percent: int
    Completeness_Indicator: str
    Valid_Day_Count: int
    Required_Day_Count: int
    Exceptional_Data_Count: int
    Null_Data_Count: int
    Primary_Exceedance_Count: float
    Secondary_Exceedance_Count: float
    Certification_Indicator: str
    Num_Obs_Below_MDL: int
    Arithmetic_Mean: float
    Arithmetic_Standard_Dev: float
    first_Max_Value: float
    first_Max_DateTime: datetime
    second_Max_Value: float
    second_Max_DateTime: datetime
    third_Max_Value: float
    third_Max_DateTime: datetime
    fourth_Max_Value: float
    fourth_Max_DateTime: datetime
    ninety_ninth_Percentile: float
    ninety_eigth_Percentile: float
    ninety_seventh_Percentile: float
    ninetieth_Percentile: float
    seventy_fifth_Percentile: float
    fiftieth_Percentile: float
    tenth_Percentile: float
    Local_Site_Name: str
    Address: str
    State_Name: str
    County_Name: str
    City_Name: str
    CBSA_Name: str
```

```python
pd.options.mode.copy_on_write = True

column_names = [column for column in clean_df]
column_names_redone = [name.replace(" ", "_") for name in column_names]

manually_renamed_columns = ["tenth_Percentile", "fiftieth_Percentile", "seventy_fifth_Percentile",
                           "ninetieth_Percentile", "ninety_seventh_Percentile", "ninety_eigth_Percentile",
                           "ninety_ninth_Percentile", "fourth_Max_DateTime", "fourth_Max_Value",
                           "third_Max_DateTime", "third_Max_Value", "second_Max_DateTime", "second_Max_Value",
                           "first_Max_DateTime", "first_Max_Value"]
for i in range(15):
    column_names_redone[-1 * i - 7] = manually_renamed_columns[i]

rename_dict = {}
for i in range(len(column_names)):
    rename_dict[column_names[i]] = column_names_redone[i]
    
placeholder = clean_df.rename(columns=rename_dict)
temp_df = placeholder.head(10000)

NaN_columns = ["Pollutant_Standard", "Local_Site_Name", "CBSA_Name", "Method_Name", "City_Name"]
NaT_columns = ["first_Max_DateTime", "second_Max_DateTime", "third_Max_DateTime", "fourth_Max_DateTime"]

temp_df[NaT_columns] = temp_df[NaT_columns].fillna(pd.NaT)
temp_df[NaN_columns] = temp_df[NaN_columns].fillna("")
temp_df[NaT_columns]
```

```python
dict_data = temp_df.to_dict(orient='records')

temp_count = 0

for datapoint in dict_data:
    param = Data(**datapoint)
    try:
        param.validate()
        print(f"{param} is valid.")
    except Exception as e:
        print(f"{param.State_Code} is invalid: {e}")
        temp_count += 1
print(temp_count)
```
